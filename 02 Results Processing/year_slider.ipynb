```python
import ipywidgets as widgets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def compare_days(day_index):

    flex_percent = 35

    '''Filenames'''
    folder1 = f'Results1/{flex_percent}/'
    folder2 = f'Results2/{flex_percent}/'

    # Filenames
    power_file1 = f'{folder1}power.csv'
    power_file2 = f'{folder2}power.csv'

    '''Process data'''

    # TYPICAL DAY
    start_ind = day_index * 24 + 2
    end_ind = start_ind + 24

    power_df1 = pd.read_csv(power_file1)
    power_df1 = power_df1.iloc[start_ind: end_ind, 1:-1]
    power_df1 /= 1000 # MW to GW
    power_df1.columns = power_df1.columns.str.replace('[_QLD]', '', regex=True) # Fix headings
    cumulative1 = power_df1.cumsum(axis=1)

    power_df2 = pd.read_csv(power_file2)
    power_df2 = power_df2.iloc[start_ind: end_ind, 1:-1]
    power_df2 /= 1000 # MW to GW
    power_df2.columns = power_df2.columns.str.replace('[_QLD]', '', regex=True) # Fix headings
    cumulative2 = power_df2.cumsum(axis=1)

    load_df1 = pd.read_csv('Load_data1.csv')
    load_df1.iloc[:, -1]

    load_df2 = pd.read_csv('Load_data2.csv')
    load_df2.iloc[:, -1]

    load_df1 /= 1000  # MW to GW
    load_df2 /= 1000  # MW to GW

    load_df1.index -= start_ind
    load_df2.index -= start_ind

    load_df1 = load_df1.iloc[start_ind: end_ind, :]
    load_df2 = load_df2.iloc[start_ind: end_ind, :]


    # labels = ['Day Peak Coordinated Charging Profile', 'Day and Night Peaks Coordinated Charging Profile']

    # plt.xlabel('Hour')
    # plt.xticks(range(0, 24, 2))
    # plt.ylabel('Load (GW)')



    colors = ['yellow', 'limegreen', 'teal', 'pink', 'crimson', 'magenta', 'orange', 'darkviolet', 'lime', 'blue', 'cyan']

    '''Set up figure'''
    # Create the figure and axes
    fig, axs = plt.subplots(1, 2, figsize=(16, 7))

    for i in range(power_df1.shape[1]): 
        if i == 0:
            axs[0].fill_between(power_df1.index-start_ind, cumulative1.iloc[:, i], 
                                alpha=0.5, label=power_df1.columns[i], 
                                color=colors[i])
        else:
            axs[0].fill_between(power_df1.index-start_ind, cumulative1.iloc[:, i], 
                                cumulative1.iloc[:, i-1], alpha=0.5, 
                                label=power_df1.columns[i], color=colors[i])


    axs[0].set_title(f'QLD Generation on April {day_index-90}, 2050 - Day Peak')
    axs[0].set_xlabel('Hour')
    axs[0].set_ylabel('Power Generated (GW)')
    axs[0].plot(load_df1.index, load_df1, color='black')
    
    # axs[0].set_ylim(0, 25)
    # axs[0].legend(loc='upper left', bbox_to_anchor=(1, 1))

    
    for i in range(power_df2.shape[1]): 
        if i == 0:
            axs[1].fill_between(power_df2.index-start_ind, cumulative2.iloc[:, i], 
                                alpha=0.5, label=power_df2.columns[i], 
                                color=colors[i])
        else:
            axs[1].fill_between(power_df2.index-start_ind, cumulative2.iloc[:, i], 
                                cumulative2.iloc[:, i-1], alpha=0.5, 
                                label=power_df2.columns[i], color=colors[i])

    axs[1].set_title(f'QLD Generation on April {day_index-90}, 2050 - Day and Night Peaks')
    axs[1].set_xlabel('Hour')
    axs[1].set_ylabel('Power Generated (GW)')
    axs[1].plot(load_df2.index, load_df2, color='black')
    # axs[1].set_ylim(0, 25)
    axs[1].legend(loc='upper left', bbox_to_anchor=(1, 1))


    # Set the overall title for the figure
    # fig.suptitle(f'{flex_percent} Percent Coordinated Charging - Day and Night Peaks', fontsize=24)
    plt.subplots_adjust(right=0.8) # make room for typical day legend

    # Display the figure
    plt.show()

def update_fig(slider_value):
    for day_index in list(range(91, 121)):
        if slider_value == day_index:
            compare_days(day_index)

slider = widgets.IntSlider(min=91, max=120, step=1, value=91, description='Slider')
slider.layout = widgets.Layout(width='100%')
widgets.interact(update_fig, slider_value=slider)
```
